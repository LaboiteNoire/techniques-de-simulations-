```python
import customtkinter
```

```python
customtkinter.set_default_color_theme("theme_0002.json")
```

```python
class App(customtkinter.CTk):
    def __init__(self,title="Modeles A et B",resolution="1200x400"):
        super().__init__()
        self.title(title)
        self.geometry("1200x800")
        self.frame_01 = customtkinter.CTkFrame(master=self,fg_color="#E5E5E5",corner_radius=0,width=900,height=240) #"#525252"
        self.frame_01.place(x=0, y=20)

        self.frame_02 = customtkinter.CTkFrame(master=self,fg_color="#E5E5E5",corner_radius=0,width=900,height=240) #"#525252"
        self.frame_02.place(x=0, y=270)

        self.frame_02 = customtkinter.CTkFrame(master=self,fg_color="#E5E5E5",corner_radius=0,width=260,height=490) #"#525252"
        self.frame_02.place(x=920, y=20)
```

```python
app = App()
app.mainloop()
```
